```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import pandas as pd
import numpy as np
from tabulate import tabulate
```

```python
nltk.download('vader_lexicon')
```

```python
#data=pd.read_csv("/content/drive/MyDrive/2021_ori.csv")
sia=SentimentIntensityAnalyzer()
```

```python
def classify_sentiment(score):
  if score>0.05:
    return 'Positive'
  elif score<=-.05:
    return 'Negative'
  else:
    return 'Neutral'
```

```python
# Apply sentiment analysis to the original dataset and create a new column 'Sentiment'
data['Sentiment'] = data['If there is anything else you would like to tell us that has not been covered by the survey questions, please use this space to do so.'].apply(lambda x: classify_sentiment(sia.polarity_scores(x)['compound']) if isinstance(x, str) else np.nan)
```

```python
data.to_csv("/content/drive/MyDrive/2021_ori.csv", index=False)
```

```python
filtered_data = data[data['If there is anything else you would like to tell us that has not been covered by the survey questions, please use this space to do so.'].notnull()]

# Apply sentiment analysis to the filtered rows and create a new column 'Sentiment'
filtered_data['Sentiment'] = filtered_data['If there is anything else you would like to tell us that has not been covered by the survey questions, please use this space to do so.'].apply(lambda x: classify_sentiment(sia.polarity_scores(x)['compound']) if isinstance(x, str) else np.nan)

# Print the filtered rows and corresponding sentiment
print(filtered_data[['If there is anything else you would like to tell us that has not been covered by the survey questions, please use this space to do so.', 'Sentiment']])
```

```python
import pandas as pd

# List of your CSV file filenames
file_names = [
    "/content/drive/MyDrive/2017_ori.csv",
    "/content/drive/MyDrive/2018_ori.csv",
    "/content/drive/MyDrive/2019_ori.csv",
    "/content/drive/MyDrive/2020_ori.csv",
    "/content/drive/MyDrive/2021_ori.csv"
]

# List to store DataFrames for each file
dfs = []

# Common columns across all years (replace with your actual column names)
common_columns = ['If there is anything else you would like to tell us that has not been covered by the survey questions, please use this space to do so.','Do you currently have a mental health disorder?', 'Sentiment']

#Loop through each CSV file
for file_name in file_names:
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_name)

    # Select only the common columns
    df = df[common_columns]

    # Append the DataFrame to the list
    dfs.append(df)

# Concatenate DataFrames vertically
final_dataset = pd.concat(dfs, ignore_index=True)

# Save the final dataset to a new CSV file
final_dataset.to_csv("/content/drive/MyDrive/Sentiment_Analysis_dataset.csv", index=False)
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

values_to_check = [5.0, 6.0, 7.0, 8.0,9.0,10.0]
values_to_ignore=[0.0,1.0,2.0,3,0,4.0]

# Read your data into a DataFrame (replace "/path/to/your/data.csv" with your actual file path)
data = pd.read_csv("/content/drive/MyDrive/Sentiment_Analysis_dataset.csv")
Stable_and_MentalDisorder=data[(data['Sentiment'].isin(['Positive'])) & (data['Do you currently have a mental health disorder?'].isin(['Yes', 'Possibly', 'Maybe']))
                                   ].shape[0]

UnStable_and_MentalDisorder=data[(data['Sentiment'].isin(['Negative', 'Neutral'])) & (data['Do you currently have a mental health disorder?'].isin(['Yes', 'Possibly', 'Maybe']))
                                   ].shape[0]

print(Stable_and_MentalDisorder)
print(UnStable_and_MentalDisorder)

#Data is representative of the fact that people willing to share are more in number and has nothing to relate with correlation of willingness and mental disorders
```

```python
labels=['Stability and Disorders', "Instability and Disorders"]
sizes=[Stable_and_MentalDisorder, UnStable_and_MentalDisorder]
colors = ['skyblue', 'lightcoral']
plt.pie(sizes, labels=labels,autopct='%1.1f%%',colors=colors, startangle=140)
plt.axis('equal')# circle
plt.title('Effect of Stability on Mental Disorders')
plt.show()
```
